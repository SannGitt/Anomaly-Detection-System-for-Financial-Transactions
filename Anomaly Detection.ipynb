**Univariate Analysis**

**3 Sigma and Mean**

**A :  Statistal Modeling For Sales**



```python
mean_sales = df['Sales'].mean()
sigma_sales = df['Sales'].std()
three_sigma_sales = 3*sigma_sales

threshold_sales_value = mean_sales + three_sigma_sales
print('Threshold Sales:', threshold_sales_value)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

sns.distplot(df['Sales'])
plt.axvspan(threshold_sales_value, df['Sales'].max(), facecolor='r', alpha=0.3)
plt.title("Sales Distribution with Outlier Region");
```

```python
sales_outliers_df = df['Sales'][df['Sales'] > threshold_sales_value]
print('Total Sales Outliers:', len(sales_outliers_df))
sales_outliers_sorted = sales_outliers_df.sort_values(ascending=False)
sales_outliers_sorted.head(5)
```

```python
(df.loc[sales_outliers_sorted.index.tolist()][['City', 'Category', 'Sub-Category', 'Product Name',
                                              'Sales', 'Quantity', 'Discount', 'Profit']]).head(10)
```

**B : Statistical Model for Profit**

```python
#Upper Limit Threshold for Profit
mean_profit = df['Profit'].mean()
sigma_profit = df['Profit'].std()
three_sigma_profit = 3 * sigma_profit

threshold_profit_upper_limit = mean_profit + three_sigma_profit
threshold_profit_lower_limit = mean_profit - three_sigma_profit

# Outlier Regions
plt.figure(figsize=(12, 6))
sns.distplot(df['Profit'])
plt.axvspan(threshold_profit_upper_limit, df['Profit'].max(), facecolor='r', alpha=0.3, label='Upper Outlier Region')
plt.axvspan(df['Profit'].min(), threshold_profit_lower_limit, facecolor='r', alpha=0.3, label='Lower Outlier Region')
plt.title('Profit Distribution with Outlier Regions')
plt.legend()
plt.show()

#  Sort Outliers
profit_outliers_df = df[(df['Profit'] > threshold_profit_upper_limit) | (df['Profit'] < threshold_profit_lower_limit)]
profit_outliers_sorted = profit_outliers_df.sort_values(by='Profit', ascending=False)
print('Total Profit Outliers:', len(profit_outliers_sorted))

# Top 10 Outlier Transactions
top_profit_outliers = profit_outliers_sorted.head(10)
top_profit_outliers



```

```python
#  Bottom 10 Outlier Transactions
bottom_profit_outliers = profit_outliers_sorted.tail(10)
bottom_profit_outliers
```

**Isolation Forest**

**A : Isolation Model trained on Sales**

```python
from sklearn.ensemble import IsolationForest

sales_ifmodel = IsolationForest(n_estimators=100,
                                contamination=0.01)
sales_ifmodel.fit(df[['Sales']])
```

```python
xx=np.linspace(df['Sales'].min(),df['Sales'].max(),len(df)).reshape(-1,1)
anomaly_score = sales_ifmodel.decision_function(xx)
outlier = sales_ifmodel.predict(xx)
plt.figure(figsize=(12, 6))
plt.plot(xx, anomaly_score, label='anomaly score')
plt.fill_between(xx.T[0], np.min(anomaly_score), np.max(anomaly_score),where=outlier==-1, color='r',alpha=.4, label='outlier region')
plt.legend()
plt.ylabel('anomaly score')
plt.xlabel('Sales');
```

```python
outlier_predictions = sales_ifmodel.predict(df[['Sales']])

sales_outliers_df = df[['Sales']]
sales_outliers_df['Outlier'] = outlier_predictions
sales_outliers_df = sales_outliers_df[sales_outliers_df['Outlier'] == -1]['Sales']

print('Total Sales Outliers:', len(sales_outliers_df))
sales_outliers_sorted = sales_outliers_df.sort_values(ascending=False)
sales_outliers_sorted.head(5)
```

**B : Isolation Forest trained on Profit**

```python
from sklearn.ensemble import IsolationForest

# Initialize and Train Model
profit_if_model = IsolationForest(n_estimators=100, contamination=0.01)
profit_if_model.fit(df[['Profit']])

# Visualize Outlier Regions
xx = np.linspace(df['Profit'].min(), df['Profit'].max(), len(df)).reshape(-1, 1)
anomaly_score = profit_if_model.decision_function(xx)
outlier = profit_if_model.predict(xx)

plt.figure(figsize=(12, 6))
plt.plot(xx, anomaly_score, label='Anomaly Score')
plt.fill_between(xx.T[0], np.min(anomaly_score), np.max(anomaly_score), where=outlier==-1, color='r', alpha=0.4, label='Outlier Region')
plt.legend()
plt.xlabel('Profit')
plt.ylabel('Anomaly Score')
plt.title('Isolation Forest - Outlier Detection for Profit')
plt.show()

# Filter and Sort Outliers
outlier_predictions = profit_if_model.predict(df[['Profit']])
profit_outliers_df = df[outlier_predictions == -1]

# View Top 10 Outlier Transactions
top_profit_outliers_if = profit_outliers_df.sort_values(by='Profit', ascending=False).head(10)
print('Top 10 Outlier Transactions (Isolation Forest):')
top_profit_outliers_if

print('Total Profit Outliers:', len(sales_outliers_df))

```

```python
# View Bottom 10 Outlier Transactions
bottom_profit_outliers_if = profit_outliers_df.sort_values(by='Profit').head(10)
print('\nBottom 10 Outlier Transactions (Isolation Forest):')
bottom_profit_outliers_if
```
